**Toy dataset = breast_Cancer**

```python
from sklearn.datasets import load_breast_cancer
data = load_breast_cancer()
data
```

```python
X = data.data
y = data.target
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.2,random_state = 2)
```

**this is implementation of Adaboost,1st of the boosting algorithms** 

```python
from sklearn.ensemble import AdaBoostClassifier
from sklearn.metrics import accuracy_score
from sklearn.model_selection import cross_val_score
```

```python
ada = AdaBoostClassifier()
np.mean(cross_val_score(ada,X,y,scoring = 'accuracy',cv = 5))
```

```python
ada.fit(X_train,y_train)
y_pred = ada.predict(X_test)
result = accuracy_score(y_test,y_pred)
result
```

**tuning the model to get the best parameters**

```python
from sklearn.model_selection import GridSearchCV
grid = {}
grid['n_estimators'] = [50,75,100,125,150,200]
grid['learning_rate'] = [0.0001, 0.001, 0.01, 0.1, 1.0]
grid['algorithm'] = ['SAMME', 'SAMME.R']

tune = GridSearchCV(ada,param_grid = grid,n_jobs = -1,cv = 5 , scoring = 'accuracy')
result = tune.fit(X_train,y_train)
result.best_score_
```
