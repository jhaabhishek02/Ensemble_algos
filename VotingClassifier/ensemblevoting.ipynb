```python
from sklearn import datasets
data = datasets.load_wine()
data
```

```python
X = data.data
y = data.target
```

```python
y.shape
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
```

```python
lr = LogisticRegression()
svm = SVC()
knn = KNeighborsClassifier()
nb = GaussianNB()
```

```python
estimators = [ ('SVM',svm) , ('KNN_ALGO',knn) , ('naive_bayes',nb)]
from sklearn.model_selection import cross_val_score
```

```python
for i in estimators:
    result = cross_val_score(i[1],X,y,cv=10,scoring = 'accuracy') 
    print(i[0],np.round(np.mean(result),2))
```

```python
from sklearn.ensemble import VotingClassifier
v_obj = VotingClassifier(estimators = estimators)
voting_result = cross_val_score(v_obj,X,y,cv=10,scoring = 'accuracy')
np.round(np.mean(voting_result),2)

```

```python
for i in range(1,4):
    for j in range(1,4):
        for k in range(1,4):
            v_obj = VotingClassifier(estimators = estimators,weights = [i,j,k])
            voting_result = cross_val_score(v_obj,X,y,cv=10,scoring = 'accuracy')
            print("i={},j={},k={}".format(i,j,k),np.round(np.mean(voting_result),2))

                
    
```
